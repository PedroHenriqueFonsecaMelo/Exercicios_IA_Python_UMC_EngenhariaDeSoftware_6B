```python
import numpy as np
```

```python
# Criando um array unidimensional
arr1 = np.array([1, 2, 3, 4])
print("Array 1D:", arr1)

# Criando um array bidimensional (matriz)
arr2 = np.array([[1, 2], [3, 4], [5, 6]])
print("Array 2D:")
print(arr2)
```

```python
# Somando arrays de mesmo tamanho
arr3 = np.array([5, 6, 7, 8])
soma = arr1 + arr3
print("Soma dos arrays:", soma)

soma = arr3 + 10
print("Soma dos arrays:", soma)
```

```python
# Multiplicação de arrays elemento por elemento
multiplicacao = arr1 * arr3
print("Multiplicação dos arrays:", multiplicacao)

multiplicacao = 2 * arr3
print("Multiplicação dos arrays:", multiplicacao)
```

```python
# Produto de duas matrizes
matriz1 = np.array([[1, 2], [3, 4]])
matriz2 = np.array([[5, 6], [7, 8]])

produto_matrizes = np.dot(matriz1, matriz2)
print("Produto de matrizes:")
print(produto_matrizes)
```

```python
# Soma de duas matrizes de mesmo tamanho
matriz_soma = matriz1 + matriz2
print("Soma de matrizes:")
print(matriz_soma)
```

```python
# Calculando a transposta de uma matriz
transposta = matriz1.T
print("Transposta da matriz:")
print(transposta)
```

```python
# Produto escalar de dois vetores (dot product)
vetor1 = np.array([1, 2, 3])
vetor2 = np.array([4, 5, 6])

produto_escalar = np.dot(vetor1, vetor2)
print("Produto escalar dos vetores:", produto_escalar)
```

```python
# Importando a biblioteca pandas
import pandas as pd
```

```python

# Criando um DataFrame com dados de exemplo
# Usamos um dicionário para passar os dados para o pandas
dados = {
    'nome': ['Alice', 'Bob', 'Carlos', 'Diana', 'Eduardo'],
    'idade': [23, 35, 30, 45, 40],
    'cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Curitiba', 'Porto Alegre']
}

# Criando o DataFrame a partir do dicionário
tabela = pd.DataFrame(dados)

# Exibindo a tabela criada
print("Tabela criada:")
print(tabela)

# Calculando a média das idades
# A função .mean() é usada para calcular a média de uma coluna específica (no caso, 'idade')
media_idades = tabela['idade'].mean()

# Exibindo a média das idades
print("\nMédia das idades:", media_idades)
```

```python
import matplotlib.pyplot as plt
```

```python

# Exemplo de dados de produtos
produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
precos = [10, 20, 15, 30, 25]  # Preço unitário de cada produto
quantidade = [500, 300, 800, 450, 600]  # Quantidade de cada produto em estoque

# Calculando o valor total de cada produto (preço * quantidade)
valor_total = [p * q for p, q in zip(precos, quantidade)]

# Criando o gráfico de barras para representar os valores individuais
plt.bar(produtos, valor_total, color='skyblue')

# Adicionando título e rótulos aos eixos
plt.title("Valor Total de Estoque por Produto")
plt.xlabel("Produtos")
plt.ylabel("Valor Total em Estoque (R$)")

# Exibindo o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Gerando dados para o gráfico
np.random.seed(0)
x = np.random.rand(50)  # 50 pontos para o eixo X
y = np.random.rand(50)  # 50 pontos para o eixo Y
z = np.random.rand(50)  # 50 pontos para o eixo Z

# Criando uma figura e um eixo 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Criando o gráfico de dispersão 3D
ax.scatter(x, y, z, color='b', marker='o')

# Adicionando título e rótulos aos eixos
ax.set_title("Gráfico de Dispersão 3D")
ax.set_xlabel("Eixo X")
ax.set_ylabel("Eixo Y")
ax.set_zlabel("Eixo Z")

# Exibindo o gráfico
plt.show()
```
