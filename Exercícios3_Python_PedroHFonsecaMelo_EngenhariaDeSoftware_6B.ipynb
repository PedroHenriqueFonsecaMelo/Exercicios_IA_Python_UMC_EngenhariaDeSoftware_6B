Exercícios de Numpy

```python
import numpy as np

# 1. Crie um array unidimensional de 50 elementos contendo números inteiros aleatórios entre 10 e 100.
array1 = np.random.randint(10, 101, 50)
print(array1)
```

```python
# 2. Gere uma matriz 4x4 com números inteiros de 1 a 16 e extraia a última linha.
matriz2 = np.arange(1, 17).reshape(4, 4)
ultima_linha = matriz2[-1]
print(ultima_linha)
```

```python
# 3. Crie um array de zeros com comprimento 15 e substitua os valores da 5ª à 10ª posição por 1.
array3 = np.zeros(15, dtype=int)
array3[4:10] = 1
print(array3)
```

```python
# 4. Mediana e moda de um array aleatório
array4 = np.random.randint(1, 101, 100)
mediana4 = np.median(array4)

valores, contagens = np.unique(array4, return_counts=True)
moda4 = valores[np.argmax(contagens)]

print(mediana4)
print(moda4)
```

```python
# 5. Substituir múltiplos de 3 por -3
array5 = np.random.randint(1, 101, 20)
for i in range(len(array5)):
    if array5[i] % 3 == 0:
        array5[i] = -3
print(array5)
```

```python
# 6. Soma das colunas de uma matriz 5x5 aleatória
matriz6 = np.random.rand(5, 5)
soma_colunas6 = matriz6.sum(axis=0)
print(soma_colunas6)
```

```python
# 8. Matriz 3x3 elevada ao quadrado
matriz8 = np.random.randint(1, 11, (3, 3))
matriz8_quadrado = matriz8 ** 2
print(matriz8_quadrado)
```

```python
# 9. Normalização de um array
array9 = np.random.rand(10)
array9_normalizado = (array9 - array9.min()) / (array9.max() - array9.min())
print(array9_normalizado)
```

```python
# 10. Matriz identidade com diagonal modificada
matriz10 = np.eye(6)
np.fill_diagonal(matriz10, np.arange(1, 7))
print(matriz10)
```

```python
# 11. Valor mais frequente em um array
array11 = np.random.randint(1, 51, 100)

# Cálculo manual da moda
valores, contagens = np.unique(array11, return_counts=True)
valor_mais_frequente = valores[np.argmax(contagens)]

print(valor_mais_frequente)
```

```python
# 12. Matriz 4x4 com diagonal substituída por 7
matriz12 = np.random.randint(1, 20, (4, 4))
np.fill_diagonal(matriz12, 7)
print(matriz12)
```

```python
# 13. Soma cumulativa dos 10 primeiros pares
array13 = np.arange(2, 21, 2)
soma_cumulativa13 = np.cumsum(array13)
print(soma_cumulativa13)
```

```python
# 14. Média das diferenças entre dois arrays
array14a = np.random.rand(15)
array14b = np.random.rand(15)
diferenca14 = array14a - array14b
media_diferenca14 = np.mean(diferenca14)
print(media_diferenca14)
```

```python
# 15. Substituir valores maiores que 50 por 50
array15 = np.random.randint(1, 101, 20)
array15[array15 > 50] = 50
print(array15)
```

```python
# 16. Reorganizar array unidimensional em matriz 3x4
array16 = np.arange(12)
matriz16 = array16.reshape(3, 4)
print(matriz16)
```

```python
# 17. Multiplicar matriz 4x4 por suas posições
matriz17 = np.random.randint(1, 10, (4, 4))
indices = np.indices((4, 4))
matriz17 = matriz17 * (indices[0] * indices[1])
print(matriz17)
```

```python
# 18. Substituir múltiplos de 5 por 0
array18 = np.random.randint(1, 101, 10)
for index in range(len(array18)):
    if array18[index] % 5 == 0:
        array18[index] = 0

print(array18)
```

```python
# 20. Encontrar o mínimo de cada linha em uma matriz
matriz20 = np.random.randint(1, 101, (5, 5))
minimo_por_linha20 = np.min(matriz20, axis=1)
print(minimo_por_linha20)
```

Exercícios de Pandas

```python
import pandas as pd
import numpy as np

dados = {
    "Produto": ["Notebook", "Mouse", "Teclado", "Monitor", "Cadeira"],
    "Preço": [3500, 15, 200, 1200, 60],
    "Quantidade": [10, 50, 30, 15, 20],
    "Categoria": ["Eletrônicos", "Acessórios", "Acessórios", "Eletrônicos", "Móveis"],
    "Desconto": [0, 0, 0, 0, 0],
    "Total": [0, 0, 0, 0, 0],
}

df = pd.DataFrame(dados)
print("1:", df, "\n")
df.to_excel("vendas.xlsx", index=False)
```

```python
# 2. Ler vendas.xlsx e exibir as 10 primeiras linhas
df2 = pd.read_excel("vendas.xlsx")
print("2:", df2.head(10), "\n")
```

```python
# 3. Filtrar produtos com preço maior que 100
linhas_filtradas = []
for index, row in df.iterrows():
    if row["Preço"] > 100:
        linhas_filtradas.append(row)
df3 = pd.DataFrame(linhas_filtradas)

# Exibindo o DataFrame filtrado
print("3:", df3, "\n")
```

```python
# 4. Adicionar uma coluna de desconto (10% para todos os produtos)
df["Desconto"] = df["Preço"] * 0.10

df.to_excel("vendas.xlsx", index=False)
print("4:", df, "\n")
```

```python
# 5. Remover linhas com valores nulos
df.dropna(inplace=True)

df.to_excel("vendas.xlsx", index=False)
print("5:", df, "\n")
```

```python
# 6. Ordenar o DataFrame pela coluna Produto em ordem alfabética
df6 = df.sort_values("Produto")
print("6:", df6, "\n")
```

```python
# 7. Criar coluna Total (Preço * Quantidade)
df["Total"] = df["Preço"] * df["Quantidade"]
print("7:", df, "\n")
```

```python
# 8. Agrupar por Categoria e calcular total de produtos
df_grouped = df.groupby("Categoria")["Quantidade"].sum().reset_index()
print("8:", df_grouped, "\n")
```

```python
# 9. Exportar o DataFrame para CSV
df.to_csv("resultados.csv", index=False)
```

```python
# 10. Ler um CSV e renomear colunas
df_csv = pd.read_csv("resultados.csv")
df_csv.rename(columns={"Produto": "Item", "Preço": "Valor"}, inplace=True)
print("10:", df_csv.tail(), "\n")
```

```python
# 11. Preencher valores nulos de uma coluna com a mediana
mediana = df["Preço"].median()
df.fillna({"Preço":mediana}, inplace=True)

df.to_excel("vendas.xlsx", index=False)
print("11: \n", df, "\n")
```

```python
# 12. Verificar colunas com valores duplicados
df12 = df[df.duplicated()]
print("12:", df12, "\n")
```

```python
# 13. Criar tabela dinâmica (soma de Quantidade por Categoria)
df13 = df.pivot_table(values="Quantidade", index="Categoria", aggfunc="sum")
print("13:", df13, "\n")
```

```python
# 14. Mesclar dois DataFrames com base no ID
dfA = pd.DataFrame({"ID": [1, 2, 3], "Nome": ["Ana", "Pedro", "Carlos"]})
dfB = pd.DataFrame({"ID": [1, 2, 3], "Idade": [25, 30, 35]})
df14 = pd.merge(dfA, dfB, on="ID", how="inner")
print("14:", df14, "\n")
```

```python
# 15. Criar DataFrame com dados faltantes e interpolar
df15 = pd.DataFrame({"Tempo": [1, 2, 3, 4, 5], "Valor": [10, np.nan, 30, np.nan, 50]})
df15["Valor"] = df15["Valor"].interpolate()
print("15:", df15, "\n")
```

```python
# 16. Criar gráfico de barras usando pandas
df.plot(x="Produto", y="Quantidade", kind="bar")
```

```python
# 17. Adicionar coluna Imposto (5% sobre Total)
df["Imposto"] = df["Total"] * 0.05

# Atualizar o arquivo
df.to_excel("vendas.xlsx", index=False)
print("17:", df, "\n")
```

```python
# 18. Exibir produtos que começam com "B"
df18 = df[df["Produto"].str.startswith("B")]
print("18:", df18, "\n")
```

```python
# 19. Criar DataFrame e calcular média das colunas numéricas
df19 = pd.DataFrame({
    "A": [10, 20, 30],
    "B": [5, 15, 25],
    "C": ["X", "Y", "Z"]
})
media19 = df19.select_dtypes(include=[np.number]).mean()
print("19:", media19, "\n")
```

```python
# 20. Agrupar por Categoria e calcular a mediana da Quantidade
df20 = df.groupby("Categoria")["Quantidade"].median().reset_index()
print("20:", df20, "\n")
```

4 Exercícios de Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Gráfico de linha para y = x^3 no intervalo de -10 a 10
x = np.linspace(-10, 10)
y = x**3
plt.figure()
plt.plot(x, y)
plt.title("y = x^3")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
# 2. Gráfico de dispersão para 100 pontos aleatórios com cores baseadas em y
np.random.seed(42)
x = np.random.rand(100)
y = np.random.rand(100)
colors = y
plt.figure()
plt.scatter(x, y, c=colors, cmap="viridis")
plt.colorbar(label="Valor de y")
plt.title("Gráfico de Dispersão")
plt.show()
```

```python
# 3. Histograma para 500 números inteiros entre 0 e 100
data = np.random.randint(0, 100, 500)
plt.figure()
plt.hist(data, bins=20, color="blue", edgecolor="black")
plt.title("Histograma")
plt.xlabel("Valor")
plt.ylabel("Frequência")
plt.show()
```

```python
# 4. Gráfico de barras para 5 categorias
categorias = ["A", "B", "C", "D", "E"]
valores = [10, 20, 15, 30, 25]
plt.figure()
plt.bar(categorias, valores, color=["red", "green", "blue", "purple", "orange"])
plt.title("Gráfico de Barras")
plt.show()
```

```python
# 5. Gráfico de pizza para representar setores de mercado
setores = ["Tecnologia", "Saúde", "Financeiro", "Varejo"]
percentuais = [25, 35, 20, 20]
plt.figure()
plt.pie(percentuais, labels=setores, autopct="%1.1f%%", colors=["blue", "red", "green", "orange"])
plt.title("Participação no Mercado")
plt.show()
```

```python
# 6. Dois gráficos de linha em subplots diferentes
x = np.linspace(0.1, 5, 100)
y1 = np.exp(x)
y2 = np.log(x)

fig, axs = plt.subplots(2, 1, figsize=(6, 8))
axs[0].plot(x, y1, color="blue")
axs[0].set_title("y = e^x")
axs[1].plot(x, y2, color="red")
axs[1].set_title("y = log(x)")
plt.show()
```

```python
# 7. Gráfico de dispersão com tamanho dos pontos variando
x = np.random.rand(100)
y = np.random.rand(100)
sizes = y * 300
plt.figure()
plt.scatter(x, y, s=sizes, alpha=0.5)
plt.title("Gráfico de Dispersão com Variação de Tamanho")
plt.show()
```

```python
# 8. Gráfico de linha com intervalo de confiança
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure()
plt.plot(x, y, color="black")
plt.fill_between(x, y - 0.2, y + 0.2, color="gray", alpha=0.3)
plt.title("y = sin(x) com Intervalo de Confiança")
plt.show()
```

```python
# 9. Adicionar título, rótulos e grade personalizada
x = np.linspace(0, 10, 100)
y = np.sqrt(x)
plt.figure()
plt.plot(x, y, label="y = sqrt(x)")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title("Gráfico com Grade Personalizada")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

```python
# 10. Três gráficos de linha no mesmo plot
x = np.linspace(0, 10, 100)
plt.figure()
plt.plot(x, np.sin(x), label="sin(x)", linestyle="--", color="blue")
plt.plot(x, np.cos(x), label="cos(x)", linestyle="-.", color="red")
plt.plot(x, np.tan(x), label="tan(x)", linestyle=":", color="green")
plt.ylim(-2, 2)
plt.legend()
plt.title("Várias Funções no Mesmo Gráfico")
plt.show()
```

```python
# 11. Gráfico de barras empilhadas
categorias = ["A", "B", "C"]
grupo1 = [10, 20, 15]
grupo2 = [5, 10, 10]
grupo3 = [8, 15, 12]

plt.figure()
plt.bar(categorias, grupo1, label="Grupo 1")
plt.bar(categorias, grupo2, bottom=grupo1, label="Grupo 2")
plt.bar(categorias, grupo3, bottom=np.array(grupo1) + np.array(grupo2), label="Grupo 3")
plt.legend()
plt.title("Gráfico de Barras Empilhadas")
plt.show()
```

```python
# 12. Gráfico de linha com função quadrática e limites ajustados
x = np.linspace(-5, 5, 100)
y = x**2
plt.figure()
plt.plot(x, y, color="purple")
plt.xlim(-5, 5)
plt.ylim(0, 25)
plt.title("Função Quadrática")
plt.show()
```

```python
# 13. Gráfico de dispersão 3D
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)
ax.scatter(x, y, z)
ax.set_title("Gráfico de Dispersão 3D")
plt.show()
```

```python
# 14. Gráfico de barras horizontal
categorias = ["A", "B", "C", "D"]
valores = [10, 15, 20, 25]

plt.figure()
plt.barh(categorias, valores, color="teal")
plt.title("Gráfico de Barras Horizontal")
plt.show()
```

```python
# 15. Gráfico da função y = 1/x
x = np.linspace(1, 100, 100)
y = 1 / x
plt.figure()
plt.plot(x, y)
plt.title("y = 1/x")
plt.show()
```

```python
# 16. Gráfico com rótulo no ponto x=5
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure()
plt.plot(x, y)
plt.annotate("Ponto de interesse", xy=(5, np.sin(5)), xytext=(5, -1))
plt.title("Gráfico com Rótulo")
plt.show()
```

```python
# 18. Gráfico de barras com valores positivos e negativos
categorias = ["A", "B", "C", "D"]
valores = [10, -15, 20, -10]
cores = ["green" if v > 0 else "red" for v in valores]
plt.figure()
plt.bar(categorias, valores, color=cores)
plt.title("Valores Positivos e Negativos")
plt.show()
```

```python
# 19. Gráfico com duas séries e legenda
x = np.linspace(0, 10, 100)
plt.figure()
plt.plot(x, np.sin(x), label="Seno", color="blue")
plt.plot(x, np.cos(x), label="Cosseno", color="red")
plt.legend()
plt.title("Seno e Cosseno")
plt.show()
```
